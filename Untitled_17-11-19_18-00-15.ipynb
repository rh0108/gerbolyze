````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
img = cv2.imread('/home/user/toys/gerbimg/foo_coarse.png')
````

````python
SCALE_FACTOR = 4

img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
w, h = img_gray.shape
img_big = cv2.resize(img_gray, (w*SCALE_FACTOR, h*SCALE_FACTOR), cv2.INTER_CUBIC)
ret, img_thr = cv2.threshold(img_big, 127, 255, 0)
img_cont_out, contours, hierarchy = cv2.findContours(img_thr, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_TC89_KCOS)
img_cont = cv2.cvtColor(img_big, cv2.COLOR_GRAY2RGB)
conts = cv2.drawContours(img_cont, contours, -1, (0, 0, 255))
````

````python
plt.imshow(img_cont, aspect='auto')
````

Plot contour length histogram
--------------

Use log-log for easier viewing as we're observing some sort of power law here as is common with random graphs. A funny fact about this distribution is that we can actually see the spot size having an effect on this. There are discrete peaks every 25 or 30 contour points that are caused by a single contour most commonly containing a discrete number of spots and every spot being in the same order of magnitude in size having roughly the same number of contour vertices.

````python
import collections
import math
cnt = collections.Counter(len(c) for c in contours)
cnt_sorted = sorted(cnt.items())
n = 100
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 3))
ax1.plot([k for k, _ in cnt_sorted[:n]], [math.log10(k) for _, k in cnt_sorted[:n]])
ax2.plot([math.log10(k) for k, _ in cnt_sorted], [math.log10(k) for _, k in cnt_sorted])
````

Count interior and exterior contours, respectively
-------------------------

The fourth element of the hierarchy tuples is the countour's parent. As we've told ```findContours``` to use the ```CCOMP``` retrieval node, we'll get $-1$ for all white contours and some positive index for all internal, black contours.

````python
np.sum(hierarchy[:,:,3] == -1), np.sum(hierarchy[:,:,3] != -1)
````

Build a dict to resolve parents to their children
-----

Structure: 
```python
{
    int(parent_index):
        (parent_contour_ndarray, [child_contour_ndarrays...])
}
```

````python
parents_and_children = {}

for i, (_1, _2, _3, parent) in enumerate(hierarchy[0]):
    if parent == -1:
        parents_and_children[i] = (contours[i], [])

for i, (_1, _2, _3, parent) in enumerate(hierarchy[0]):
    if parent != -1:
        parent_cont, child_set = parents_and_children[parent]
        child_set.append(contours[i])
````

How many children (holes) does a contour usually have?
---

````python
import collections
import math
chld = collections.Counter(len(children) for parent, children in parents_and_children.values())
chld_sorted = sorted(chld.items())[:20]
plt.plot([k for k, _ in chld_sorted[:n]], [math.log10(k) for _, k in chld_sorted[:n]])
````

````python
def write_csv(scale, filename):
    with open(filename, 'w') as f:
        print('"Index","X (mm)","Y (mm)","Arc Angle (Neg = CW)"', file=f)
        for parent, _children in parents_and_children.values():
            for x, y in parent[:,0]:
                print('"{}","{}","{}",""'.format(i, scale*float(x), scale*float(y)), file=f)
````

````python
def svg_format_curve(scale, curve):
    mid = ('L{} {} '.format(x, y) for x, y in curve)
    return ' '.join(['M{} {}'.format(*curve[0]), *mid, 'Z'])

def write_svg(scale, filename):
    with open(filename, 'w') as f:
        f.write('<svg width="'+str(w*scale*SCALE_FACTOR)+'" height="'+str(h*scale*SCALE_FACTOR)+'" xmlns="http://www.w3.org/2000/svg">')
        for parent, children in parents_and_children.values():
            f.write('<path d="{}"/>'.format(' '.join(
                        svg_format_curve(scale, c[:,0]) for c in [parent, *children]
                    )))
        f.write('</svg>')
````
